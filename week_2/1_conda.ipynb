# Part 1 — Setting Up Our Python Environment with Conda

## Why Do We Need an Environment?

In data science, we rely on external libraries such as:

- `numpy`
- `pandas`
- `matplotlib`
- `seaborn`

Different projects may require:
- Different Python versions  
- Different versions of the same library  
- Different combinations of dependencies  

If we install everything globally on our system:
- Projects can interfere with each other  
- Updating one package may break another project  
- Reproducibility becomes difficult  

To avoid these problems, we use **isolated environments**.

---

## What Is Conda?

Conda is both:

- A package manager (like `pip`)
- An environment manager

It allows us to:

- Install Python
- Install libraries
- Create isolated environments
- Reproduce setups across different machines

In this course, we will create a **local environment inside our project folder**.

This means:
- Everything stays contained inside the project directory
- Nothing affects your global Python installation
- You can delete the folder to remove everything cleanly

---

# Step 1 — Create a Local Environment

We will create an environment called:

`venv`

And we will place it inside the current working directory using the `--prefix` option.

The `--prefix` flag tells conda:

"Create the environment in this exact folder."

---

## Open a Terminal (Not Inside Jupyter)

Navigate to your project folder.

Then run:

```bash
conda create --prefix ./venv python=3.12 -y

### What This Command Does

- `conda create` → create a new environment  
- `--prefix ./venv` → create it locally in a folder called `venv`  
- `python=3.12` → install Python version 3.12  
- `-y` → automatically confirm installation  

After running this command, you should see a new folder:

```
venv/
```


This folder now contains:

- A full Python installation  
- Its own `site-packages` directory  
- All dependencies required to run Python independently  

It is completely isolated from your system-wide Python installation.

---

## Step 2 — Activate the Environment

After creating the environment, we must activate it before installing packages or running Python.

Activation tells your terminal:

"Use the Python interpreter and packages inside this folder."

### macOS / Linux
```bash
conda activate ./venv
```

### Windows
```bash
conda activate .\venv
```

After activation, your terminal prompt may change to indicate you are now working inside the `venv` environment.
For example, it might look like this:

```
(venv) user@computer:~/project$
```
This means that any Python commands you run will now use the Python interpreter and libraries from the `venv` environment.

---
## Step 3 — Install Packages
Now that we have our environment activated, we can install the packages we need for our data science projects.
For example, to install `numpy`, `pandas`, `matplotlib`, and `seaborn`, run:

```bash
pip install numpy pandas matplotlib seaborn
```

This command will install the specified libraries into the `venv` environment without affecting your global Python installation.
You can verify that the packages are installed by running:

```bash
pip list
```

This will show you a list of all the packages installed in the current environment, confirming that your libraries are ready to use for your data science projects.

---
## Step 4 — Deactivate the Environment

When you are done working in the `venv` environment, you can deactivate it to return to your global Python environment.

To deactivate the environment, simply run:

```bash
conda deactivate
```
This command will switch you back to your system's default Python environment, and your terminal prompt will return to normal, indicating that you are no longer working inside the `venv` environment.
You can reactivate the `venv` environment at any time by running the activation command again.

---
## Summary
- We created a local environment called `venv` using conda.
- We activated the environment to use its isolated Python interpreter and libraries.
- We installed necessary data science packages without affecting our global Python installation.
- We can deactivate the environment when we're done to return to our global Python setup.
This setup allows us to manage dependencies effectively and ensures that our projects remain reproducible and isolated from each other.
