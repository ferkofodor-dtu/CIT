# Part 4 — Data Visualization with Matplotlib

So far, we have:

- Created numerical arrays with NumPy
- Manipulated tabular data with Pandas
- Computed statistics and filtered datasets

Now we move to **visualization**.

Visualization is critical in data science because:

- It helps us detect patterns
- It reveals outliers
- It exposes structure that summary statistics may hide
- It improves communication of results

We will begin with **Matplotlib**, the foundational plotting library in Python.

---

## Importing Matplotlib

Matplotlib is typically imported as follows:

```python
import matplotlib.pyplot as plt
```

When working inside Jupyter, we enable inline plotting:

```python
%matplotlib inline
```

This ensures plots appear directly inside the notebook.

---

# 1. Basic Line Plot

We start with a simple example.

```python
import numpy as np

x = np.arange(0, 10)
y = x ** 2

plt.plot(x, y)
plt.show()
```

This creates a basic line plot.

---

## Adding Labels and Titles

Good visualizations must be clearly labeled.

```python
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values (x squared)")
plt.show()
```

### Note: Always label your axes.

---

# 2. Scatter Plot

Scatter plots are useful for showing relationships between variables.

```python
plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

Scatter plots are particularly important in data science because they reveal:

- Trends
- Clusters
- Outliers
- Non-linear relationships

---

# 3. Plotting Data from a DataFrame

Let us load our CSV file and visualize it.

```python
import pandas as pd

df = pd.read_csv("data/myfile.csv")
df.head()
```

## Scatter Plot: Study Hours vs Grade

```python
plt.scatter(df["study_hours"], df["grade"])
plt.xlabel("Study Hours")
plt.ylabel("Grade")
plt.title("Study Hours vs Grade")
plt.show()
```

Discussion:

- Do students who study more tend to get higher grades?
- Are there any outliers?

---

# 4. Bar Plot

Bar plots are useful for categorical comparisons.

```python
plt.bar(df["student"], df["grade"])
plt.xticks(rotation=45)
plt.title("Grades per Student")
plt.xlabel("Student")
plt.ylabel("Grade")
plt.show()
```

Note:

`rotation=45` makes the student names readable.

---

# 5. Histogram

Histograms show distributions.

```python
plt.hist(df["grade"], bins=5)
plt.title("Distribution of Grades")
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.show()
```

Discussion:

- Is the grade distribution symmetric?
- Are most grades clustered in a particular range?

---

# 6. Multiple Plots in One Figure

Sometimes we want multiple visualizations together.

```python
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.scatter(df["study_hours"], df["grade"])
plt.title("Scatter")

plt.subplot(1, 2, 2)
plt.hist(df["grade"], bins=5)
plt.title("Histogram")

plt.tight_layout()
plt.show()
```

This creates two plots side by side.
