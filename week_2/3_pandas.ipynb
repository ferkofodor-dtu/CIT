# Part 3 — Introduction to Pandas

In this section, we move from numerical arrays to **structured data** using Pandas.

NumPy is ideal for numerical computation.  
Pandas is designed for **tabular data**, similar to spreadsheets or SQL tables.

Pandas provides two main data structures:

- `Series` → one-dimensional labeled data  
- `DataFrame` → two-dimensional labeled table  

Most real-world datasets are handled using DataFrames.

---

## Importing Pandas

```python
import pandas as pd
```

---

# 1. Creating a Pandas Series

A Series is similar to a NumPy array, but with labels (index).

```python
s = pd.Series([10, 20, 30, 40])
s
```

Notice the index on the left.

You can also specify your own index.

```python
s = pd.Series([10, 20, 30], index=["a", "b", "c"])
s
```

Access elements using labels:

```python
s["b"]
```

---

# 2. Creating a DataFrame

A DataFrame is a table with rows and columns.

## From a Dictionary

```python
data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "score": [85, 90, 95]
}

df = pd.DataFrame(data)
df
```

Each key becomes a column.

---

# 3. Inspecting Data

Pandas provides helpful inspection tools.

```python
df.head(2)
```

```python
df.tail(2)
```

```python
df.info()
```

```python
df.describe()
```

- `head(N)` shows the first N rows  
- `tail(N)` shows the last N rows
- `info()` shows column types and missing values  
- `describe()` computes summary statistics  

---

# 4. Selecting Data

## Selecting a Column

```python
df["age"]
```

## Selecting Multiple Columns

```python
df[["name", "score"]]
```

## Selecting Rows with loc (label-based)

```python
df.loc[0]
```

## Selecting Rows with iloc (position-based)

```python
df.iloc[1]
```

---

# 5. Filtering Data

We can filter rows using conditions.

```python
df[df["age"] > 28]
```

This selects only rows where age is greater than 28.

---

# 6. Adding and Modifying Columns

```python
df["passed"] = df["score"] > 88
df
```

You can also modify values:

```python
df["age"] = df["age"] + 1
df
```

---

# 7. Handling Missing Data

Let’s introduce missing values.

```python
df.loc[1, "score"] = None
df
```

Check for missing values:

```python
df.isna()
```

```python
df.isna().sum()
```

Drop missing values:

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

By default, `dropna()` removes any row with at least one missing value. You can specify `axis=1` to drop columns instead.

Or fill them:

```python
df["score"] = df["score"].fillna(df["score"].mean())
df
```

---

# 8. Grouping Data

Grouping is very common in data analysis.

```python
data = {
    "department": ["IT", "IT", "HR", "HR", "Finance"],
    "salary": [50000, 60000, 45000, 48000, 70000]
}

df2 = pd.DataFrame(data)
df2
```

```python
df2.groupby("department")["salary"].mean()
```

This computes the average salary per department.
