# Combining Datasets and Feature Engineering

## Goal

We now move from analyzing datasets individually to **combining them** to create new insights.

---

## Why combine datasets?

Each dataset provides only part of the picture:

- population → number of people  
- GDP → total economic output  
- life expectancy → health / living conditions  

To answer meaningful questions, we need to **bring them together**.

---

## Key Question

> Do richer countries live longer?

---

## Step 1 — Merge datasets

We combine the datasets using:

- `country`
- `year`

This ensures we compare values that belong together.

---

## Important checks

Before and after merging:

- consistent country codes  
- matching time ranges  
- missing values  

---

## Step 2 — Feature Engineering

We create a new variable:

> **GDP per capita = GDP / population**

This allows fair comparison between countries of different sizes.

---

## Why this matters

- total GDP favors large countries  
- GDP per capita reflects **wealth per person**
