# Introduction to Time Series Data

---

# 1. What is Time Series Data?

A **time series** is a sequence of data points collected **over time**, usually at regular intervals.

Examples:

- daily temperature measurements  
- monthly airline passengers  
- hourly electricity consumption  
- stock prices over time  

---

## Key characteristics of time series

Time series data is different from regular tabular data because:

- **Order matters** → data points are not independent  
- **Time is an explicit variable**  
- Observations are often **correlated with past values**

---

## Common patterns in time series

### 1. Trend
Long-term increase or decrease in the data.

Example:
- increasing bike usage over years

---

### 2. Seasonality
Repeating patterns at fixed intervals.

Examples:
- higher usage in summer than winter  
- daily peaks (morning vs evening)

---

### 3. Noise
Random fluctuations that are not part of a pattern.

---

### 4. Cycles (optional)
Longer-term variations that are not strictly periodic.

---

## Why time series are important

Time series data appears in many real-world problems:

- transportation demand  
- weather forecasting  
- energy consumption  
- finance  
- user behavior  

In this course, we will focus on:

- **visualizing time patterns**
- **extracting features from time**
- **understanding temporal behavior**

### Part 4 - Seasonal features

Seasonal features can capture patterns that repeat over specific time intervals, such as daily, weekly, or yearly cycles. In this part, we will create new features that represent the seasonality in the data. One such feature is the time of day, which can capture workday vs. weekend patterns, while another feture could be the season of the year (spring, summer, fall, winter) based on the month. These features can help us to better understand and model the temporal patterns in bike usage.

````python
def map_month_to_season(month):
    if month in [3, 4, 5]:
        return 1  # Spring
    elif month in [6, 7, 8]:
        return 2  # Summer
    elif month in [9, 10, 11]:
        return 3  # Fall
    else:
        return 4  # Winter

df['season'] = df['month'].apply(map_month_to_season)
df[['month', 'season']].head(10)
````

### Part 5 - Lagging features (optional)

Lagging features are created by shifting the target variable by a certain number of time steps. This can help capture temporal dependencies in the data, such as how past bike usage might influence future usage. For example, we could create a lag feature that represents the number of bikes rented in the previous hour or day. This can be particularly useful for time series forecasting models.

Have a look at the function below:

````python
def create_lag_feature(df, lag_hours, target_col='count'):
    df[f'lag_{lag_hours}h'] = df[target_col].shift(lag_hours)
    return df

df = create_lag_feature(df, lag_hours=1, target_col='casual')
df[['datetime', 'casual', 'lag_1h']].head(10)
````
