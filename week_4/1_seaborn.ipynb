## Why compare Matplotlib and Seaborn?

**Matplotlib** is a general-purpose plotting library. It gives a lot of control, but often requires more code.

**Seaborn** is built on top of Matplotlib. It is designed for **statistical visualization** and usually produces cleaner plots with less code.

# Example 1: Scatter plot
## Concept

A scatter plot is useful when we want to visualize the relationship between two numerical variables.

We will use the `iris` dataset and compare:

`sepal_length`

`sepal_width`

### Matplotlib version:

````python
plt.figure(figsize=(6, 4))
plt.scatter(iris["sepal_length"], iris["sepal_width"])
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title("Matplotlib: Sepal length vs Sepal width")
plt.show()
````

### Seaborn version:

````python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width")
plt.title("Seaborn: Sepal length vs Sepal width")
plt.show()
````

### Side by side comparison:

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].scatter(iris["sepal_length"], iris["sepal_width"])
axes[0].set_xlabel("Sepal length")
axes[0].set_ylabel("Sepal width")
axes[0].set_title("Matplotlib")

sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", ax=axes[1])
axes[1].set_title("Seaborn")

plt.tight_layout()
plt.show()
````

### Extension: add categories

This is where Seaborn becomes especially convenient.

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Matplotlib
species_list = iris["species"].unique()
for species in species_list:
    subset = iris[iris["species"] == species]
    axes[0].scatter(subset["sepal_length"], subset["sepal_width"], label=species)

axes[0].set_xlabel("Sepal length")
axes[0].set_ylabel("Sepal width")
axes[0].set_title("Matplotlib with categories")
axes[0].legend()

# Seaborn
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species", ax=axes[1])
axes[1].set_title("Seaborn with hue")

plt.tight_layout()
plt.show()
````

# Example 5: Box plot
## Concept

A box plot helps compare distributions across categories and identify outliers.

We will compare **total_bill** across different days in the `tips` dataset.

### Matplotlib version:

````python
days = ["Thur", "Fri", "Sat", "Sun"]
data_by_day = [tips[tips["day"] == day]["total_bill"] for day in days]

plt.figure(figsize=(6, 4))
plt.boxplot(data_by_day, tick_labels=days)
plt.xlabel("Day")
plt.ylabel("Total bill")
plt.title("Matplotlib: Total bill by day")
plt.show()
````

### Seaborn version:

````python
plt.figure(figsize=(6, 4))
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Seaborn: Total bill by day")
plt.show()
````

# Example 7: Heatmap
## Concept

A heatmap is a graphical representation of data where values are represented by colors. It is particularly useful for visualizing correlation matrices or any tabular data.

The `flights` dataset is excellent for this after pivoting.

````python
flights_matrix = flights.pivot(index="month", columns="year", values="passengers")
flights_matrix
````

### Matplotlib version:

````python
plt.figure(figsize=(10, 5))
plt.imshow(flights_matrix, aspect="auto")
plt.colorbar(label="Passengers")
plt.xticks(range(len(flights_matrix.columns)), flights_matrix.columns, rotation=90)
plt.yticks(range(len(flights_matrix.index)), flights_matrix.index)
plt.title("Matplotlib: Flights heatmap")
plt.show()
````

### Seaborn version:

````python
plt.figure(figsize=(10, 5))
sns.heatmap(flights_matrix, cmap="viridis")
plt.title("Seaborn: Flights heatmap")
plt.show()
````

# Summary of plot types

| Plot type    | Main use                                     | Good dataset       |
| ------------ | -------------------------------------------- | ------------------ |
| Scatter plot | relationship between two numerical variables | `iris`, `penguins` |
| Line plot    | trends over time / ordered variable          | `flights`          |
| Bar plot     | compare categories                           | `tips`             |
| Histogram    | distribution of one variable                 | `penguins`, `tips` |
| Box plot     | compare distributions and outliers           | `tips`, `penguins` |
| Violin plot  | compare distributions with density           | `tips`, `penguins` |
| Heatmap      | matrix-style values                          | `flights`          |
| Pairplot     | multi-variable exploration                   | `iris`, `penguins` |
