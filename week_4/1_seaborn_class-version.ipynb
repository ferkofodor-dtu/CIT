# Seaborn vs Matplotlib: A Visual Introduction

## Lesson goal

In this lesson, we will compare **Matplotlib** and **Seaborn** through a series of examples.  
The main idea is to:

1. introduce a plot type with **Matplotlib**,
2. recreate the same idea with **Seaborn**,
3. discuss what Seaborn makes easier,
4. practice on real datasets that come with Seaborn.

## Datasets Used in the Examples

Below are the datasets used in the demonstrations. These datasets are included in **Seaborn** and can be loaded using `sns.load_dataset()`.

| Dataset | Description |
|---|---|
| `iris` | Measurements of iris flowers (sepal and petal size) for three species. Commonly used for classification and visualization examples. |
| `flights` | Monthly airline passenger counts from 1949 to 1960. Useful for time-series plots and heatmaps. |
| `penguins` | Physical measurements of penguins (bill size, flipper length, body mass) from different species and islands. |
| `tips` | Restaurant data including bill amounts, tips, day, time, and party size. Often used for categorical comparisons and distribution plots. |

These are useful because they contain both **numerical** and **categorical** variables, which makes them ideal for data visualization.

## Why compare Matplotlib and Seaborn?

**Matplotlib** is a general-purpose plotting library. It gives a lot of control, but often requires more code.

**Seaborn** is built on top of Matplotlib. It is designed for **statistical visualization** and usually produces cleaner plots with less code.

# Example 1: Scatter plot
## Concept

A scatter plot is useful when we want to visualize the relationship between two numerical variables.

We will use the `iris` dataset and compare:

`sepal_length`

`sepal_width`

### Matplotlib version:

````python
plt.figure(figsize=(6, 4))
plt.scatter(iris["sepal_length"], iris["sepal_width"])
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title("Matplotlib: Sepal length vs Sepal width")
plt.show()
````

### Seaborn version:

````python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width")
plt.title("Seaborn: Sepal length vs Sepal width")
plt.show()
````

### Side by side comparison:

````python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

axes[0].scatter(iris["sepal_length"], iris["sepal_width"])
axes[0].set_xlabel("Sepal length")
axes[0].set_ylabel("Sepal width")
axes[0].set_title("Matplotlib")

sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", ax=axes[1])
axes[1].set_xlabel("Sepal length")
axes[1].set_ylabel("Sepal width")
axes[1].set_title("Seaborn")

plt.tight_layout()
plt.show()
````

### Extension: add categories

This is where Seaborn becomes especially convenient.

````python
# creating a custom colormap with 3 colors

import matplotlib

cmap = matplotlib.cm.get_cmap('vanimo', 3)
cmap
````

Indexing the species_list in two ways together with the color palette

````python
species_list = iris["species"].unique()
species_list

for i in range(len(species_list)):
    spe = species_list[i]
    color = cmap(i)
    print(i, spe, color)

print('----------------------')
for idx, spe in enumerate(species_list):
    print(idx, spe, cmap(idx))
````

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

cmap = matplotlib.cm.get_cmap('vanimo', 3)

# Matplotlib
species_list = iris["species"].unique()
for idx, species in enumerate(species_list):
    subset = iris[iris["species"] == species]
    axes[0].scatter(subset["sepal_length"], subset["sepal_width"], label=species, color=cmap(idx), zorder=10, edgecolor='lime', s=100)
    

axes[0].set_xlabel("Sepal length")
axes[0].set_ylabel("Sepal width")
axes[0].set_title("Matplotlib with categories")
axes[0].legend()

# Seaborn
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species", ax=axes[1], zorder=10, edgecolor='lime', s=100, palette='spring')
axes[1].set_title("Seaborn with hue")

axes[0].grid(axis='y', color='red', alpha=0.5, linestyle='--')
axes[1].grid(axis='both', color='gray', alpha=0.9, linewidth=5, zorder=0)


plt.tight_layout()
plt.show()
````

# Example 2: Line plot
## Concept

A line plot is useful to show trends across an ordered variable such as time.

We will use the `flights` dataset.

````python
flights
````

````python
flights_pivot = flights.pivot(index="year", columns="month", values="passengers")
flights_yearly = flights.groupby("year", as_index=False)["passengers"].mean()
flights_yearly.head()
````

````python
flights_pivot
````

### Matplotlib version:

````python
plt.figure(figsize=(7, 4))

plt.plot(flights_yearly["year"], flights_yearly["passengers"], marker="*", markersize=20, color='red', markerfacecolor='gold', markeredgecolor='lime')
plt.xlabel("Year")
plt.ylabel("Average passengers")
plt.title("Matplotlib: Average passengers per year")
plt.show()
````

## Using two plots to separate the line plot and the markers

````python
plt.plot(flights_yearly["year"], flights_yearly["passengers"], color='blue', zorder=0)
plt.scatter(flights_yearly["year"], flights_yearly["passengers"], marker="*", color='red', s=200, zorder=1)
````

### Seaborn version:

````python
plt.figure(figsize=(7, 4))
sns.lineplot(data=flights_yearly, x="year", y="passengers", marker="*", markersize=20, color='red', markerfacecolor='gold', markeredgecolor='lime')
plt.title("Seaborn: Average passengers per year")
plt.show()
````

### Side-by-side comparison:

````python
fig, ax = plt.subplots(1, 2, figsize=(10, 3))

# Matplotlib
years = flights['year'].unique()
for yr in years:
    df_temp = flights[flights['year'] == yr]
    ax[0].plot(df_temp["month"], df_temp["passengers"], label=f'{yr}')
ax[0].legend(loc='upper right')

# Seaborn
sns.lineplot(data=flights, x='month', y='passengers', hue='year', ax=ax[1])
````

````python
# Your code goes here
df1 = flights[flights['year'] == 1949]
df2 = flights[flights['year'] == 1960]

fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=True)

axes[0].plot(df1["month"], df1["passengers"], marker="*", markersize=20, color='red', markerfacecolor='gold', markeredgecolor='lime')

axes[0].set_xlabel("Month")
axes[0].set_ylabel("Passengers")
axes[0].set_title("1949")

# Seaborn
sns.lineplot(data=df2, x="month", y="passengers", marker="*", markersize=20, color='orange', markerfacecolor='C2', markeredgecolor='gray', ax=axes[1])
axes[1].set_title("1960")


for ax in axes:
    ax.grid(axis='y', color='gray', alpha=0.5, linestyle='--')
    ax.set_ylim(50, 650)

plt.tight_layout()
plt.show()
````

# Example 3: Bar plot
## Concept

A bar plot is useful for comparing values across categories.

We will compare the **average total bill by day** using the `tips` dataset.

Prepare aggregated data for Matplotlib

````python
tips.head()
````

````python
avg_bill_by_day = tips.groupby('day')['total_bill'].mean().reset_index()
avg_bill_by_day
````

### Matplotlib version:

````python
plt.figure(figsize=(6, 4))
plt.bar(x=avg_bill_by_day["day"], height=avg_bill_by_day["total_bill"])
plt.xlabel("Day")
plt.ylabel("Average total bill")
plt.title("Matplotlib: Average total bill by day")
plt.show()
````

### Seaborn version:

````python
plt.figure(figsize=(6, 4))
sns.barplot(data=avg_bill_by_day, x="day", y="total_bill")
plt.title("Seaborn: Average total bill by day")
plt.show()
````

### Side-by-side comparison

````python
# Your code goes here

avg_bill_by_day = tips.groupby(['day', 'sex'])['total_bill'].mean().reset_index()


fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].bar(avg_bill_by_day["day"], avg_bill_by_day["total_bill"])

axes[0].set_xlabel("Day")
axes[0].set_ylabel("Total Bill")
axes[0].set_title("MPL")

# Seaborn
sns.barplot(data=avg_bill_by_day, x="day", y="total_bill", hue='sex', ax=axes[1])
axes[1].set_title("Seaborn")
axes[1].legend(loc='lower right')


for ax in axes:
    ax.grid(axis='y', color='gray', alpha=0.5, linestyle='--')

plt.tight_layout()
plt.show()
````

# Example 4: Histogram
## Concept

A histogram shows the distribution of a single numerical variable.

We will use **bill_length_mm** from the `penguins` dataset.

Clean missing values

````python
penguins.head()
````

````python
penguins_clean = penguins.dropna()
penguins_clean = penguins_clean[['bill_length_mm']]
penguins_clean
````

### Matplotlib version:

````python
plt.figure(figsize=(6, 4))
plt.hist(penguins_clean["bill_length_mm"], bins=30)
plt.xlabel("Bill length (mm)")
plt.ylabel("Frequency")
plt.title("Matplotlib: Histogram of bill length")
plt.show()
````

````python
plt.figure(figsize=(6, 4))
sns.histplot(data=penguins_clean, x="bill_length_mm", bins=30)
plt.title("Seaborn: Histogram of bill length")
plt.show()
````

### Side-by-side comparison

````python
# Your code goes here
# MPL: body_mass_g	
# SNS: flipper_length_mm

fig, ax = plt.subplots(1,2, figsize=(12, 4))
data = penguins.dropna()[['body_mass_g', 'flipper_length_mm', 'sex']]

for s in data['sex'].unique():
    ax[0].hist(data[data['sex']==s]['body_mass_g'], alpha=0.6, label=s)

ax[0].set_xlabel('Body Mass (g)')
ax[0].legend()

sns.histplot(data=data, x='flipper_length_mm', hue='sex', ax=ax[1], alpha=0.8, palette='summer')
````

# Example 5: Box plot
## Concept

A box plot helps compare distributions across categories and identify outliers.

We will compare **total_bill** across different days in the `tips` dataset.

### Matplotlib version:

````python
days = ["Thur", "Fri", "Sat", "Sun"]
data_by_day = [tips[tips["day"] == day]["total_bill"] for day in days]

fig, ax = plt.subplots(1,2, figsize=(12, 3))
ax[0].boxplot(data_by_day, tick_labels=days)

avg = data_by_day[0].mean()
std = data_by_day[0].std()

ax[1].hist(data_by_day[0])
ax[1].axvline(avg, color='red')
ax[1].axvline(avg - 2 * std, color='gray')
ax[1].axvline(avg + 2 * std, color='gray')
ax[1].set_xlabel('Thursday')

plt.show()
````

### Seaborn version:

````python
tips.head()
````

````python
plt.figure(figsize=(6, 4))
sns.boxplot(data=tips, x="day", y="total_bill", hue='smoker')
plt.title("Seaborn: Total bill by day")
plt.show()
````

# Example 7: Heatmap
## Concept

A heatmap is a graphical representation of data where values are represented by colors. It is particularly useful for visualizing correlation matrices or any tabular data.

The `flights` dataset is excellent for this after pivoting.

````python
flights_matrix = flights.pivot(index="month", columns="year", values="passengers")
flights_matrix
````

### Matplotlib version:

````python
plt.figure(figsize=(10, 5))
plt.imshow(flights_matrix, aspect="auto", cmap='jet')
plt.colorbar(label="Passengers")
plt.xticks(range(len(flights_matrix.columns)), flights_matrix.columns, rotation=90)
plt.yticks(range(len(flights_matrix.index)), flights_matrix.index)
plt.title("Matplotlib: Flights heatmap")
plt.show()
````

### Seaborn version:

````python
plt.figure(figsize=(10, 5))
sns.heatmap(flights_matrix, cmap="jet", annot=True, annot_kws={"size": 10}, fmt=".0f")
plt.title("Seaborn: Flights heatmap")
plt.show()
````

# Summary of plot types

| Plot type    | Main use                                     | Good dataset       |
| ------------ | -------------------------------------------- | ------------------ |
| Scatter plot | relationship between two numerical variables | `iris`, `penguins` |
| Line plot    | trends over time / ordered variable          | `flights`          |
| Bar plot     | compare categories                           | `tips`             |
| Histogram    | distribution of one variable                 | `penguins`, `tips` |
| Box plot     | compare distributions and outliers           | `tips`, `penguins` |
| Violin plot  | compare distributions with density           | `tips`, `penguins` |
| Heatmap      | matrix-style values                          | `flights`          |
| Pairplot     | multi-variable exploration                   | `iris`, `penguins` |
