# Interconnected Exercise Chain (A → B → C)

These exercises build on each other.  
The output you create in A will be saved to disk and then reused in B and C.

You should end this chain with a plot and a short written interpretation.

You will create a synthetic dataset about students:
- their exam scores across multiple tests
- and their study behavior
Then you will engineer features and visualize patterns.

Use the file path `data/synthetic_scores.csv` throughout.

## A) NumPy: Generate the dataset and save it

Goal: generate a realistic-looking dataset using NumPy and save it as a CSV file.

### A1. Create reproducible random data

1. Set a random seed (so everyone gets the same numbers).
2. Create `n = 200` students.
3. Generate these arrays:

- `student_id`: integers from 1 to n
- `study_hours`: random integers between 0 and 20 (inclusive)
- `sleep_hours`: random floats between 4 and 9
- `test1`, `test2`, `test3`: exam scores (0–100)

Guidance for realism:
- Make scores depend loosely on `study_hours` and `sleep_hours` plus some noise.
- Clip the final scores to the range [0, 100].

### A2. Combine into a 2D array

Combine the columns into a single NumPy matrix with shape `(n, 6)`:

`student_id, study_hours, sleep_hours, test1, test2, test3`

Check:
- `matrix.shape`
- `matrix[:5]`

### A3. Save to CSV

Save it to:

`data/synthetic_scores.csv`

Rules:
- Include a header row with the column names
- Use comma separators

Hint: `np.savetxt(...)` can do this.

Deliverable for A:
- a CSV file on disk
- and a printout of the first 5 rows loaded back using NumPy (to confirm it saved correctly)

## B) Pandas: Load the file and engineer features

Goal: load the CSV into a DataFrame and create new features that will be useful for analysis and visualization.

### B1. Load the CSV

Load `data/synthetic_scores.csv` into a DataFrame called `df`.

Check:
- `df.head()`
- `df.info()`
- `df.describe()`

### B2. Create engineered features

Add these new columns:

1. `avg_score`  
   The average of `test1`, `test2`, `test3`.

2. `score_std`  
   The standard deviation of `test1`, `test2`, `test3` (how consistent the student is).

3. `improved`  
   A boolean that is `True` if `test3 > test1`, else `False`.

4. `performance_band`  
   A categorical label based on `avg_score`:

   - `"low"` if avg_score < 50
   - `"medium"` if 50 <= avg_score < 75
   - `"high"` if avg_score >= 75

Hints:
- Use vectorized operations (no loops).
- `df[cols].mean(axis=1)` is useful.
- `pd.cut(...)` is useful for creating bands.

Deliverable for B:
- `df.head()` showing the new columns
- A count of how many students fall into each `performance_band`

- x-axis: `score_std`
- y-axis: `avg_score`
