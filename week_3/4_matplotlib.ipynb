# 1. Basic Line Plot

We start with a simple example.

````python
import numpy as np

x = np.arange(0, 10)
y = x ** 2

plt.plot(x, y)
plt.show()
````

This creates a basic line plot.

---

## Adding Labels and Titles

Good visualizations must be clearly labeled.

````python
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values (x squared)")
plt.show()
````

### Note: Always label your axes.

---

# 2. Scatter Plot

Scatter plots are useful for showing relationships between variables.

````python
plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
````

Scatter plots are particularly important in data science because they reveal:

- Trends
- Clusters
- Outliers
- Non-linear relationships

# Multiple Plots in One Figure

Sometimes we want multiple visualizations together.

````python
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.scatter(df["study_hours"], df["grade"])
plt.title("Scatter")

plt.subplot(1, 2, 2)
plt.hist(df["grade"], bins=5)
plt.title("Histogram")

plt.tight_layout()
plt.show()
````

This creates two plots side by side.

Random numbers normally change every time code runs.

By setting a **seed**, we ensure that:

- everyone generates the **same dataset**
- results are **reproducible**
- debugging and comparison become easier
